```python
from numpy.core.function_base import linspace
import matplotlib.pyplot as plt
import numpy as np
import random  
from math import sqrt, tan, exp

def fun(x):
  fx = (1- x**2)**(3/2)
  return fx
n = 1000000
suma = 0
x = np.linspace(0,1,200)
y = (1-x**2)**(3/2)

for i in range(n):
  z = random.random()
  suma = suma + fun(z)

print("El área paroximada es %.4f" %(suma/n))
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.fill_between(x,y, color = "gray")
plt.plot(x,y ,color ="black")
plt.show()
```

```python
from numpy.core.function_base import linspace
import matplotlib.pyplot as plt
import numpy as np
import random  
from math import sqrt, tan, exp

def fun(x):
  fx = exp(x+x**2)
  return fx
n = 1000000
suma = 0
x = np.linspace(-2,2,200)
y = np.exp(x+x**2)

for i in range(n):
  z = random.uniform(-2,2)
  suma = suma + fun(z)

print("El área aproximada es %.4f" %(4*suma/n))
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.fill_between(x,y, color = "red")
plt.plot(x,y ,color ="black")
plt.show()
```
